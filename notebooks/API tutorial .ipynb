```python
from yahooquery import Ticker
```

```python
import pandas as pd
import numpy as np
```

```python
petr = Ticker("PETR4.SA")
```

```python
# max period
petr.history(period="max")
```

```python
#details
petr.summary_detail
```

```python
#specific time
petr.history(start='2021-01-01', end="2021-03-05")
```

```python
#historical results
petr.income_statement()
```

```python
#historical results transposed
petr_results = petr.income_statement()# Chama Demonstração de resultados
petr_results = petr_results.transpose()           # Transpõe a matriz
petr_results.columns = petr_results.iloc[0,:] # Renomeia colunas
petr_results = petr_results.iloc[2:,1:]       # Seleciona dados
#petr_results = petr_results.iloc[:, 1::-1]    # Inverte colunas
petr_results
```
